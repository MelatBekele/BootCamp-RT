### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
gmaps.configure(api_key=g_key)
```

```python
dataNA = data.dropna()
dataNA.head()
```

```python
locations = dataNA[['Lat', 'Lng']].astype(float)
weights = dataNA['Humidity'].astype(float)

fig = gmaps.figure()
heat_Layer = gmaps.heatmap_layer(locations, weights=weights , dissipating=False, max_intensity=5,
                                 point_radius = 1)
fig.add_layer(heat_Layer)
fig
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# hotel inforamtion
hotel_loc = pd.DataFrame(dataNA, columns=["City", "Country", "Lat", "Lng"])

hotel_loc["Hotel"] = ""
hotel_loc.head()
```

```python
from prettytable import prettytable
```

```python
Hotel_loc = []

params = {
    "keyword": "hotel",
    "radius": 500,
    "type": "hotel",
    "key": g_key
}
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
for index, row in hotel_loc.iterrows():
    lat = row["Lat"]
    lng = row["Lng"]
    params["location"] = f"{lat},{lng}"

    address = requests.get(base_url, params=params)
    name_address = address.json()
    try:
        Hotel_loc.append(name_address['results'][0]['name'])
    except IndexError:
        Hotel_loc.append(np.nan)
```

```python
hotel_loc['Hotel']= hotel_loc
hotel_loc.dropna()
hotel_loc.to_csv('hotel_list.csv')
hotel_loc
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_loc.iterrows()]
locations = hotel_loc[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations)
fig.add_layer(markers)


# Display figure
fig
```
